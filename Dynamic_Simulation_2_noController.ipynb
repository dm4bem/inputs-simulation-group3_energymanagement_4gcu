# Simulation 2: Without controller

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import dm4bem
```

```python
controller = False
Kp = 1e3
explicit_Euler = True
imposed_time_step = False
Δt = 3600    # s, imposed time step   
```

```python
# MODEL
# =====
# Thermal circuits
TC = dm4bem.file2TC('TC.csv', name='', auto_number=False)

# by default TC['G']['q11'] = 0 # Kp -> 0, no controller (free-floating
if controller:
    TC['G']['q11'] = Kp     # Kp -> ∞, almost perfect controller

# State-space
[As, Bs, Cs, Ds, us] = dm4bem.tc2ss(TC)
# dm4bem.print_TC(TC)
```

```python
λ = np.linalg.eig(As)[0]    # eigenvalues of matrix As
dtmax = 2 * min(-1. / λ)    # max time step for Euler explicit stability
dt = dm4bem.round_time(dtmax)

if imposed_time_step:
    dt = Δt

dm4bem.print_rounded_time('dt', dt)
```

## Input data set

```python
# INPUT DATA SET
# ==============
input_data_set = pd.read_csv('input_data_set.csv',
                             index_col=0,
                             parse_dates=True)
#input_data_set
```

```python
input_data_set = input_data_set.resample(
    str(dt) + 'S').interpolate(method='linear')
#input_data_set.head()
```

## Input vector in time

```python
# Input vector in time from input_data_set
u = dm4bem.inputs_in_time(us, input_data_set)
#u.head()
```

## Initial conditions

```python
# Initial conditions
θ0 = 20                     # °C, initial temperatures
θ = pd.DataFrame(index=u.index)
θ[As.columns] = θ0          # fill θ with initial valeus θ0
```

## Time integration

```python
I = np.eye(As.shape[0])     # identity matrix

if explicit_Euler:
    for k in range(u.shape[0] - 1):
        θ.iloc[k + 1] = (I + dt * As) @ θ.iloc[k] + dt * Bs @ u.iloc[k]
else:
    for k in range(u.shape[0] - 1):
        θ.iloc[k + 1] = np.linalg.inv(
            I - dt * As) @ (θ.iloc[k] + dt * Bs @ u.iloc[k])
```

## Outputs

```python
# outputs
y = (Cs @ θ.T + Ds @  u.T).T
```

```python
Kp = TC['G']['q11']     # controller gain
S = 9                   # m², surface area of the toy house
q_HVAC = Kp * (u['q11'] - y['θ6']) / S  # W/m²
y['θ6']
```

## Plots

```python
data = pd.DataFrame({'To': input_data_set['To'],
                     'θi': y['θ6'],
                     'Etot': input_data_set['Etot'],
                     'q_HVAC': q_HVAC})
```

### Plots using matplotlib

Alternativelly, we may use `matplotlib`.

## Discussion

In this version of the code, following the simulation with a perfect heat controller, we decided to remove the controller. To determin if the controller is deleted, we look at the second graph, and notice that the added heat flow is non-existent.
The inside temperature depends now on the outside one. The variation of the inside temperature is increasing as the outside temperature is increasing. Nevertheless, we can see that they are not exactly the same : the variation of the inside temperature is much more smooth. This may be due to the capacity of the materials and of the air. A brief and fast peak of the outside temperature will not not influence that much the inside temperature. On top of that, this is also depending on the size of the taken time-step. 

In addition, the inside temperature is logically higher then the outside one. It can be explained by the incoming heat flow through the sun shining. The window allow this flow to heat the inside temperature, wich will cause with a certain delay, a heat flow towards the outside. This delay can be observed by overlaying the two graphs. The E, representing the sun shining, causes an instant increasing of the inside temperature, because of the transparency of the window. However, the inside temperature decreases slower because of the loss through the walls and the window (convection, conduction, radiation).
